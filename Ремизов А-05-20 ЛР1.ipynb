```python
anums = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']
aletters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
abrackets = ['(', ')']
aoper = ['+', '-', '*', '/']
aoper1 = ['+', '*', '/']
aoperH = ['*', '/']
aoperL = ['+', '-']
alphabet = anums + abrackets + aoper + aletters + [' ']
brackets_control = 0
Unominus_fl = False
bin_oper = False
alphabet_fl = False
IntChar_fl = False

def MultySymb (i, String):
    global IntChar_fl
    st=""
    if (String[i] in aletters):
        while (i<len(String) and String[i] in aletters):
            st +=(String[i])
            i+=1
    elif (String[i] in anums):
        while (i<len(String) and String[i] in anums):
            st+=(String[i])
            i+=1
    if (i<len(String) and not ((String[i] in (aoper + abrackets)) or (String[i]==" "))):
            IntChar_fl = True
    i-=1
    return st, i

def Priority(ch1, ch2):
    if (((ch1 in aoperH) and (ch2 in aoperH)) or ((ch1 in aoperL) and (ch2 in aoperL))):
        return "eq"
    elif ((ch1 in aoperH) and (ch2 in aoperL)):
        return "gr"
    elif ((ch1 in aoperL) and (ch2 in aoperH)):
        return "le"
    
def String_check(string1):
    global brackets_control, Unominus_fl, bin_oper, alphabet_fl, IntChar_fl
    #for i in range(0, len(string1)):
    i=0
    while (i<len(string1)):
        if(not(string1[i] in alphabet)):
            alphabet_fl = True
            break
        elif (string1[i] == '('):
            brackets_control+=1
        elif (string1[i] == ')'):
            brackets_control-=1
        elif (string1[i] in (anums + aletters)):
            buf, i = MultySymb(i, string1)
        elif (string1[i] in aoper1):
            if((i==0) or (string1[i-1] in aoper) or (string1[i-1] == ' ' and string1[i-2] in aoper)):
                bin_oper = True
                break
        i+=1
            
def Polskaya_zapis(string):
    global brackets_control, Unominus_fl, bin_oper, alphabet_fl, IntChar_fl
    string1 = string.lstrip()
    string1 = ' '.join(string1.split())
    print(string1)
    String_check(string1)    
    
    stack = []
    queue = []
    
    if (alphabet_fl):
        print("Ошибка! Недопустимый символ строки\n")
    elif (bin_oper):
        print("Ошибка! Проверьте знаки операций.\n")
    elif (IntChar_fl):
        print("Ошибка! Неверно введено число или переменная.\n")
    elif (brackets_control !=0):
        print("Ошибка! Неверное количество скобок.\n")
    else:
        x=""
        print("Строка введена корректно.\n")
        i = 0
        #for i in range(len(string1)):
        while (i<len(string1)):
            if (string1[i] in (anums + aletters)):
                buf, n = MultySymb(i, string1)
                queue.append(buf)
                while (n>i):
                    i+=1
            elif (string1[i] in aoper):
                if ((not stack) or (stack[len(stack)-1] == '()')):
                    stack.append(string1[i])
                elif (Priority(string1[i], stack[len(stack)-1]) == "gr"):
                    stack.append(string1[i])
                else:
                    while (i< len(string1) and len(stack)-1>=0 and not ((Priority(stack[len(stack)-1], string1[i]) == "le") or stack[len(stack)-1] == '(')):
                        queue.append(stack.pop())
                    stack.append(string1[i])
            elif (string1[i]=='('):
                stack.append(string1[i])
            elif (string1[i]==')'):
                while (stack[len(stack)-1]!='('):
                    queue.append(stack.pop())
                stack.pop()
            else:
                i+=0
            i+=1
        for i in range(len(stack)):
            queue.append(stack.pop())
        x=' '.join(queue)
        print("Польская запись: ",x)   
        return x
    print("Инфиксная запись: ", string1)
    

def Schet(string):
    global u, y, t, r, e, w, q, l, k, j, h, g, f, d, s
    stack = []
    string1 = string.lstrip()
    lst = string1.split()
    for i in range(len(lst)):
        if (lst[i] == 'u'):
            lst[i] = str(u)
        elif (lst[i] == 'y'):
            lst[i] = str(y)
        elif (lst[i] == 't'):
            lst[i] = str(t)
        elif (lst[i] == 'r'):
            lst[i] = str(r)
        elif (lst[i] == 'e'):
            lst[i] = str(e)
        elif (lst[i] == 'w'):
            lst[i] = str(w)
        elif (lst[i] == 'q'):
            lst[i] = str(q)
        elif (lst[i] == 'l'):
            lst[i] = str(l)
        elif (lst[i] == 'k'):
            lst[i] = str(k)
        elif (lst[i] == 'j'):
            lst[i] = str(j)
        elif (lst[i] == 'h'):
            lst[i] = str(h)
        elif (lst[i] == 'g'):
            lst[i] = str(g)
        elif (lst[i] == 'f'):
            lst[i] = str(f)
        elif (lst[i] == 'd'):
            lst[i] = str(d)
        elif (lst[i] == 's'):
            lst[i] = str(s)
    i = 0
    while (i<len(lst)):
        if (lst[i][0] in (anums + aletters)):
            stack.append(lst[i])
        else:
            if (lst[i]=="*"):
                buf1 = float(stack.pop())
                buf2 = float(stack.pop())
                stack.append(buf2*buf1)
            elif (lst[i]=="/"):
                buf1 = float(stack.pop())
                buf2 = float(stack.pop())
                stack.append(buf2/buf1)
            elif (lst[i]=="+"):
                buf1 = float(stack.pop())
                buf2 = float(stack.pop())
                stack.append(buf2+buf1)
            elif (lst[i]=="-"):
                buf1 = float(stack.pop())
                buf2 = float(stack.pop())
                stack.append(buf2-buf1)
        i+=1
    return stack.pop()
```
