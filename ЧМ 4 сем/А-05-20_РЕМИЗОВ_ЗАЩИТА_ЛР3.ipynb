Размер матрицы $N=(7+Q)^2$

Матрица А:
* на главной диагонали $-8N-\pi$
* на первых над и под диагоналях $N$
* на $7+Q-1$, $7+Q$, $7+Q+1$ над и под диагоналях $N$

Вектор $b$ определяется следующим образом: $b=Az$, где $z_i=Q$, $1 \leq i \leq N$


Метод решения соответствует Q:
1. $LU$ разложение. Реализовать нужно в виде двух функций: одна реализует разложение, записывая его в матрицу $А$, вторая решает систему. Создавать матрицы L и U нельзя.

```python
import math as m
import copy
q=2
N= (7+q)**2
A=[]

def AInput(A, n):
    q=m.sqrt(n)-7
    for i in range (n):
        A.append([])
        for j in range(n):
            A[i].append(0)
    
    for i in range (n):
        for j in range (n):
            if (i==j):
                A[i][j]=-8*n-m.pi
            elif ((i==0 and j==(n-1)) or(j==0 and i==n-1)):
                A[i][j]=n
            elif (abs(i-j)==7+q-1) or (abs(i-j)==7+q) or (abs(i-j)==7+q+1):
                A[i][j]=n

    return A

AInput(A, N)
print(A)

b=[]
for i in range (N):
    b.append(m.sqrt(N)-7)

print(b)
```

```python

def LU(A, n):
    L=[]
    for i in range (n):
        L.append([])
        for j in range (n):
            L[i].append(0)
           
            
    for i in range (n):
        L[i][i]=1
        for k in range (i, n-1):
            if(abs(A[k][i])>abs(A[i][i])):
                buf=A[k]
                A[k]=A[i]
                A[i]=buf
                
            M=A[k+1][i]/A[i][i]
            L[k+1][i]=M
            for j in range (n):
                A[k+1][j]=A[k+1][j]-A[i][j]*M
                
    for i in range(n):
        for j in range(n):
            if (A[i][j]==0):
                A[i][j]=copy.copy(L[i][j])
            
    return A

def Sol(A, n, b):
    A= LU(A,n)
    
    print(A, '\n'*3)
    
    y=[0]*n
    y[0]=copy.copy(b[0])
    
    for i in range (1, n):
        s=0
        for j in range(i):
            s+=y[j]*A[i][j]
        y[i]=b[i]-s
         
    x=[0]*n
    x[n-1]=y[n-1]/A[n-1][n-1]
    
    for i in range (n-2, -1, -1):
        s=0 
        for j in range (n-1, i, -1):
            s+=x[j]*A[i][j]            
            
        x[i]=(y[i]-s)/A[i][i]
    return [x]



print(Sol(A, N, b))
```
